```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
df.groupby('Order_Year')['Sales'].sum().plot(kind='line', color='blue', marker='o')
plt.title('Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')

plt.subplot(1, 2, 2)
df.groupby('Order_Year')['Profit'].sum().plot(kind='line', color='green', marker='o')
plt.title('Profit Over Time')
plt.xlabel('Year')
plt.ylabel('Profit')

plt.tight_layout()
plt.show()
```

## ðŸ“ˆ Sales & Profit Over Time: Key Trends

 - Overall sales show a significant upward trend from 2015 to 2017 after an initial dip from 2014 to 2015.

 - Profit demonstrates a consistent year-over-year increase throughout the observed period (2014-2017).

 - The growth in sales appears to accelerate considerably between 2016 and 2017.

 - While profit grows steadily, the rate of profit increase seems to be higher between 2015 and 2016 compared to 2014-2015.

```python
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
df.groupby('Segment')['Sales'].sum().sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Sales')

plt.subplot(1,2,2)
df.groupby('Segment')['Profit'].sum().sort_values(ascending=False).plot(kind='bar', color='lightgreen')
plt.title('Profit by Segment')
plt.xlabel('Segment')
plt.ylabel('Profit')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df.groupby('Ship_Mode')['Sales'].sum().sort_values(ascending=False).plot(kind='bar', color='pink')
plt.title('Sales by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Sales')
plt.show()

plt.figure(figsize=(8, 6))
df.groupby('Ship_Mode')['Profit'].sum().sort_values(ascending=False).plot(kind='bar', color='yellow')
plt.title('Profit by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Profit')
plt.show()
```

### ðŸ“Š Sales & Profit vs Quantity â€“ Insights
 - Sales vs Quantity (Left Plot)

    - High Variability: Same quantity can yield vastly different sales (e.g., Quantity = 5 â†’ Sales: $0 to $20K+), implying varying product values.

    - Outliers: High-priced items sold in small quantities create spikes in sales.

    - Non-linear Trend: Sales increase with quantity but tend to saturate at higher levels (e.g., Quantity > 10).

 - Profit vs Quantity (Right Plot)
    - No Clear Trend: Profit doesnâ€™t consistently increase with quantity.

    - High Fluctuation: Profit ranges from large losses to large gains at any quantity.

    - Costs & Discounts Impact: Varying margins and discounts likely drive the inconsistency.

    - Losses Across the Board: Negative profits appear across almost all quantity levels.

    - Possible Sweet Spot: Mid-quantities (4â€“6) show some clustering of higher profits, but not conclusive.
